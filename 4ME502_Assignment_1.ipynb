```python
# Creates a string called "text" that contains the text used for this assignment
text = """We have developed speed, but we have shut ourselves in.
Machinery that gives abundance has left us in want.
Our knowledge has made us cynical.
Our cleverness, hard and unkind. We think too much and feel too little.
More than machinery we need humanity. More than cleverness we need kindness and gentleness."""


# Function that splits sentences
def split_sentences(text):

    # Uses '.' as delimiter, splits text whenever a period is found
    sentences = text.split(r'.')

    # Removes whitespaces from before and after the sentences
    # Keeps sentence if the sentence is not an empty string
    # Returns the sentences
    return [sentence.strip() for sentence in sentences if sentence.strip()]


# Function that counts words
def count_words(text):

    # Splits the text by whitespaces to get words and saves this in a list
    words = text.split()

    # Returns the length of the list of words, same as the word count in the text
    return len(words)


# Function that orders sentences by length
def order_sentences_by_length(text):

    # Splits the text into sentences using the split_sentences function
    sentences = split_sentences(text)

    # Sorts sentences in descending order based on their respective word count
    sorted_sentences = sorted(sentences, key=count_words, reverse=True)

    # Returns the list of sentences in descending order of length
    return sorted_sentences


# Function that extracts the five longest words and sorts them alphabetically
def extract_longest_words(text):

    # Converts all characters in the text into lowercase to avoid case-issues
    words = text.lower()

    # Splits the text by whitespaces to get words
    # Removes non-alphanumeric characters (like periods and commas) from each word
    # Saves the cleaned words in a set which automatically removes any duplicates
    cleaned_words = {
        ''.join(char for char in word if char.isalnum()) for word in words.split()
    }

    # Converts the set to a list to sort it
    cleaned_words = list(cleaned_words)

    # Sorts the list of words by length (descending) and then alphabetically
    for i in range(len(cleaned_words)):
        for j in range(i + 1, len(cleaned_words)):

            # Swap words if a longer word is found
            # or if two words have equal length but are out of alphabetical order
            if (len(cleaned_words[j]) > len(cleaned_words[i])) or (
                len(cleaned_words[j]) == len(cleaned_words[i])
                and cleaned_words[j] < cleaned_words[i]):
                # Swaps the out of order words
                cleaned_words[i], cleaned_words[j] = cleaned_words[j], cleaned_words[i]

    # Extracts the five longest words from the now sorted list
    top_five = cleaned_words[:5]

    # Sorts the five longest words alphabetically before returning
    top_five.sort()

    # Returns the five longest words in alphabetical order
    return top_five


# Task 1 - Amount of Sentences
# Splits the text into a list of sentences using the split_sentences function
# Counts the amount of sentences by getting the length of the list
sentence_count = len(split_sentences(text))

# Prints the number of sentences
print(f"\nTask 1: \nThis text contains {sentence_count} sentences")


# Task 2 - Amount of Words
# Counts the amount of words in the text using the count_words function
word_count = count_words(text)

# Prints the number of sentences
print(f"\nTask 2: \nThis text contains {word_count} words")


# Task 3 - Order of Sentences by Length
# Orders sentences by their length (word count) using the order_sentences_by_length function
ordered_sentences = order_sentences_by_length(text)

# Iterates through the ordered sentences using enumerate to create a ranking starting from 1
# Prints them in order along with their amount of words
print("\nTask 3:")
for i, sentence in enumerate(ordered_sentences, start=1):
  # If i is 1, prints the longest sentence, same logic applied for each subsequent ordered sentence
  if (i == 1):
    print(f"The longest sentence in this text contains {count_words(sentence)} words")
  elif (i == 2):
    print(f"The second longest sentence in this text contains {count_words(sentence)} words")
  elif (i == 3):
    print(f"The third longest sentence in this text contains {count_words(sentence)} words")
  elif (i == 4):
    print(f"The fourth longest sentence in this text contains {count_words(sentence)} words")
  elif (i == 5):
    print(f"The fifth longest sentence in this text contains {count_words(sentence)} words")
  elif (i == 6):
    print(f"The sixth longest sentence in this text contains {count_words(sentence)} words")
    # Since we only have seven sentences in total
    # this prints the shortest sentence since i > 6 (i.e the shortest sentence)
  else:
    print(f"The shortest sentence in this text contains {count_words(sentence)} words")

# Task 4 - Five Longest Words in Alphabetical Order
# Gets the five longest words in alphabetical order using the extract_longest_words function
longest_words = extract_longest_words(text)
# Prints them in alphabetical order
print("\nTask 4: \nThe five longest words in this text ordered alphabetically are:")
print(f"{longest_words}")
```
